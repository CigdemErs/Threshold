```python
import cv2
import os
from glob import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = 'content/drive/My Drive/Colab Notebooks/SOCOFing/'
output_path = 'Threshold'

os.makedirs(output_path, exist_ok=True)
```

```python
real_output_path = os.path.join(output_path, "Real")
altered_output_path = os.path.join(output_path, "Altered")

threshold_value = 155
```

```python
# Threshold the Real fingerprints
# Create output directory if not exists
os.makedirs(real_output_path, exist_ok=True)

counter = 0 # for visualization

for image_path in glob("/content/drive/My Drive/Colab Notebooks/SOCOFing/Real/*.BMP"):
    # Load the input image.
    img = cv2.imread(image_path)
    if img is not None:
      # Convert image to grayscale
      gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

      # Threshold using Otsu's thresholding
      ret, thresh = cv2.threshold(gray, threshold_value, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

      # Save the output image
      output_image_path = os.path.join(real_output_path, os.path.basename(image_path))
      cv2.imwrite(output_image_path, thresh)

    if counter < 4:
      plt.subplot(2,2, 1)
      plt.imshow(gray, 'gray')
      plt.title('Input')
      plt.show()

      plt.subplot(2,2, 2)
      plt.imshow(thresh, 'gray')
      plt.title('Output')
      plt.show()

      counter += 1
```

```python
# Threshold the Altered-Easy fingerprints
# Create output directory if not exists
os.makedirs(altered_output_path, exist_ok=True)

counter = 0 # for visualization

for image_path in glob("/content/drive/My Drive/Colab Notebooks/SOCOFing/Altered/Altered-Easy/*.BMP"):
    # Load the input image.
    img = cv2.imread(image_path)
    if img is not None:
      # Convert image to grayscale
      gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

      # Threshold using Otsu's thresholding
      ret, thresh = cv2.threshold(gray, threshold_value, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

      # Save the output image
      output_image_path = os.path.join(altered_output_path, os.path.basename(image_path))
      cv2.imwrite(output_image_path, thresh)

      if counter < 4:
        plt.subplot(2,2, 1)
        plt.imshow(gray, 'gray')
        plt.title('Input')
        plt.show()

        plt.subplot(2,2, 2)
        plt.imshow(thresh, 'gray')
        plt.title('Output')
        plt.show()

        counter += 1
```
